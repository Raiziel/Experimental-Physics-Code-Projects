```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import integrate
from pylab import * 

t=arange(0.,10.,0.01)
# Here A is zero because we end up with divison by zero that conflicts with f0
x0=1
v0=6.
omega=2*(math.pi)
omega0=2*(math.pi)
beta=0.
f0=0.4
tau=1
delta=0



omega1=sqrt(-beta**2+omega0**2)
B1=x0-cos(delta)
B2=1/omega1*(v0-omega*sin(delta)+beta*x0-beta*cos(delta))
t2=cos(omega*t-delta)+e**(-beta*t)*(B1*cos(omega1*t)+B2*sin(omega1*t))


axes = plt.gca()
axes.set_xlim([0,10.])
axes.set_ylim([-2,2])

plt.plot(t,t2,'g-',linewidth=1.5,label='Problem 4',alpha=0.9)
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.show()
```

```python
beta=0.5
A=sqrt(f0**2)/((omega0**2-omega**2)**2+4*beta**2*omega**2)
delta=0
omega1=sqrt(-beta**2+omega0**2)
B1=x0-A*cos(delta)
B2=1/omega1*(v0-A*omega*sin(delta)+beta*x0-beta*A*cos(delta))
t2=A*cos(omega*t-delta)+e**(-beta*t)*(B1*cos(omega1*t)+B2*sin(omega1*t))


axes = plt.gca()
axes.set_xlim([0,8])
axes.set_ylim([-2,2])

plt.plot(t,t2,'g-',linewidth=2,alpha=0.9)
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```
